# QUESTIONÁRIO DE AVALIAÇÃO DO ENSINO REMOTO EMERGENCIAL 2020 - UFPA

Este questionário objetiva avaliar o Ensino Remoto Emergencial (ERE) por disciplina ministrada. Foi aplicado de forma virtual no Google Forms para os alunos do campus. Com os dados deste questionário é possível compreender as principais dificuldades nesta modalidade de ensino, a fim de melhorar o processo de ensino-aprendizagem durante a pandemia da Covid-19. O questionário é composto por cinco dimensões:  (1) Atitude profissional, (2) Gestão Didática, (3) Processo Avaliativo, (4) Avaliação das Atividades Curriculares e (5) Autoavaliação discente.

### Cada dimensão possui quesitos a serem avaliados, conforme listados abaixo:

DIMENSÃO 1: ATITUDE PROFISSIONAL:

* 1.1. Foi assíduo(a) e pontual às aulas síncronas.
* 1.2. Cumpriu a carga horária prevista nas aulas.
* 1.3. Foi disponível para atender e orientar. 
* 1.4. Interagiu com a turma com civilidade e respeito. 
* 1.5. Mostrou-se seguro no conhecimento do conteúdo 

DIMENSÃO 2: GESTÃO DIDÁTICA:

* 2.6. Apresentou o plano de ensino. 
* 2.7. Executou a programação prevista.
* 2.8. Ensinou com linguagem clara e objetiva.
* 2.9. Estimulou a participação dos estudantes. 
* 2.10. Fez a recapitulação dos assuntos que ministrou.  
* 2.11. Utilizou recursos pedagógicos adequados ao Ensino Remoto Emergencial. 
* 2.12. Orientou o desenvolvimento das atividades acadêmicas.
* 2.13. Ofereceu alternativas para garantir a aprendizagem no ERE. 

DIMENSÃO 3: PROCESSO AVALIATIVO:

* 3.14. Avaliou de forma coerente com o que foi ministrado no ERE.                
* 3.15. Avaliou com critérios definidos e expostos de forma clara. 
* 3.16. Discutiu os resultados de cada procedimento de avaliação. 

DIMENSÃO 4: AVALIAÇÃO DAS ATIVIDADES CURRICULARES:

* 4.1. Os objetivos de aprendizagem da disciplina foram alcançados. 
* 4.2. Os conteúdos ministrados foram adequados ao ERE.
* 4.3. Os conteúdos foram articulados a situações ou problemas reais. 
* 4.4. A carga horária foi suficiente para cumprir os conteúdos previstos

DIMENSÃO 5: AUTOAVALIAÇÃO DISCENTE:

* 5.1. Eu possuía os pré-requisitos necessários para o bom acompanhamento da disciplina.
* 5.2. Eu possuía a infraestrutura necessária (equipamentos, internet, mobiliário, ambiente, etc) para o bom acompanhamento da disciplina.
* 5.3. Meu estado psicológico contribuiu para o bom acompanhamento da disciplina. 
* 5.4. Dediquei o esforço necessário à disciplina. 
* 5.5. Estabeleci relação entre o conteúdo abordado e outros conteúdos já conhecidos. 
* 5.6. Participei frequentemente das aulas desta disciplina de maneira síncrona. (Legenda da frequência: 1 significa até 25%; 2 significa até 50%; 3 significa até 75% e 4 significa até 100%). 
* 5.7. Realizei e participei ativamente de todas as atividades (listas de exercícios, trabalhos, etc.) exigidas na disciplina.
* 5.8. Estou satisfeito(a) com o que aprendi na disciplina.


### A avaliação de cada quesito é realizada através da seguinte escala:

 |ESCALA   |             DESCRIÇÃO               |
|:--------:|:-------------------------------------|
|**1**  | **Não** concordo com a afirmativa.         |
|**2**  | Concordo **em parte** com a afirmativa.    |
|**3**  | Concordo **em boa parte** com a afirmativa.|
|**4**  | Concordo **totalmente** com a afirmativa.  |


para expressar as opiniões dos(as) alunos(as) sobre a qualidade para cada item, sendo que a nota 1 significa que não concorda com o quesito e a nota 4 que concorda totalmente com questão avaliada, já as notas 2 e 3 concordam em parte.
